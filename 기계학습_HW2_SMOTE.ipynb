```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import numpy as np

# 데이터 불러오기
loan_data = pd.read_csv('loan_data.csv')

# 데이터 준비 및 더미 변수 변환
X = pd.get_dummies(loan_data.drop('not.fully.paid', axis=1), columns=['purpose'], drop_first=True)
y = loan_data['not.fully.paid']

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 데이터 표준화
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 로지스틱 회귀 모델 훈련 및 평가 (SMOTE 적용 전)
model = LogisticRegression(random_state=42, max_iter=1000)
model.fit(X_train_scaled, y_train)
predictions = model.predict(X_test_scaled)
report = classification_report(y_test, predictions, output_dict=True)

# SMOTE 적용 전 결과 저장
results = pd.DataFrame({
    'Sampling_Strategy': ['Pre-SMOTE'],
    'K_Neighbors': ['N/A'],
    'Precision': [report['macro avg']['precision']],
    'Recall': [report['macro avg']['recall']],
    'F1-Score': [report['macro avg']['f1-score']],
    'Accuracy': [report['accuracy']]
})

# 하이퍼파라미터 범위 설정
sampling_strategies = [i/10 for i in range(2, 11)]
k_neighbors_options = range(1, 11)

# 하이퍼파라미터 조합에 대해 모델 훈련 및 평가 (SMOTE 적용 후)
for strategy in sampling_strategies:
    for neighbors in k_neighbors_options:
        smote = SMOTE(sampling_strategy=strategy, k_neighbors=neighbors, random_state=42)
        X_sm, y_sm = smote.fit_resample(X_train_scaled, y_train)
        model.fit(X_sm, y_sm)
        predictions = model.predict(X_test_scaled)
        report = classification_report(y_test, predictions, output_dict=True)
        results = results.append({
            'Sampling_Strategy': strategy,
            'K_Neighbors': neighbors,
            'Precision': report['macro avg']['precision'],
            'Recall': report['macro avg']['recall'],
            'F1-Score': report['macro avg']['f1-score'],
            'Accuracy': report['accuracy']
        }, ignore_index=True)

# F1-Score 그래프 출력
plt.figure(figsize=(10, 5))  # 그래프의 크기를 조정
# SMOTE 적용 전 F1-Score
pre_smote_f1 = results[results['Sampling_Strategy'] == 'Pre-SMOTE']['F1-Score'].values[0]
plt.axhline(y=pre_smote_f1, color='r', linestyle='-', label='F1 Pre-SMOTE', linewidth=2.5)

# SMOTE 적용 후 F1-Score
for k in k_neighbors_options:
    subset = results[(results['K_Neighbors'] == k) & (results['Sampling_Strategy'] != 'Pre-SMOTE')]
    plt.plot(subset['Sampling_Strategy'], subset['F1-Score'], marker='o', linestyle='-', label=f'F1 (k={k})', linewidth=2)

plt.title('F1-Score for different SMOTE hyperparameters')
plt.xlabel('Sampling Strategy')
plt.ylabel('F1-Score')
plt.xticks(np.arange(min(sampling_strategies), max(sampling_strategies) + 0.1, 0.1))
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()

# Accuracy 그래프 출력
plt.figure(figsize=(10, 5))  # 그래프의 크기를 조정
# SMOTE 적용 전 Accuracy
pre_smote_acc = results[results['Sampling_Strategy'] == 'Pre-SMOTE']['Accuracy'].values[0]
plt.axhline(y=pre_smote_acc, color='b', linestyle='--', label='Accuracy Pre-SMOTE', linewidth=2.5)

# SMOTE 적용 후 Accuracy
for k in k_neighbors_options:
    subset = results[(results['K_Neighbors'] == k) & (results['Sampling_Strategy'] != 'Pre-SMOTE')]
    plt.plot(subset['Sampling_Strategy'], subset['Accuracy'], marker='x', linestyle='--', label=f'Accuracy (k={k})', linewidth=2)

plt.title('Accuracy for different SMOTE hyperparameters')
plt.xlabel('Sampling Strategy')
plt.ylabel('Accuracy')
plt.xticks(np.arange(min(sampling_strategies), max(sampling_strategies) + 0.1, 0.1))
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()

# Display pre-SMOTE results
pre_smote_results = results[results['Sampling_Strategy'] == 'Pre-SMOTE']

# Get the best result for each Sampling_Strategy after SMOTE application
best_results_per_strategy = results[results['Sampling_Strategy'] != 'Pre-SMOTE'].groupby('Sampling_Strategy').apply(
    lambda x: x.loc[x['F1-Score'].idxmax()]
)

# Combine both pre-SMOTE and best post-SMOTE results
combined_results = pd.concat([pre_smote_results, best_results_per_strategy])

combined_results.reset_index(drop=True, inplace=True)
combined_results
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# 데이터 불러오기
loan_data = pd.read_csv('loan_data.csv')

# 데이터 준비 및 더미 변수 변환
X = pd.get_dummies(loan_data.drop('not.fully.paid', axis=1), columns=['purpose'], drop_first=True)
y = loan_data['not.fully.paid']

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 데이터 표준화
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 로지스틱 회귀 모델 훈련 (SMOTE 적용 전)
model_before_smote = LogisticRegression(random_state=42, max_iter=1000)
model_before_smote.fit(X_train_scaled, y_train)

# 테스트 세트로 평가 (SMOTE 적용 전)
predictions_before_smote = model_before_smote.predict(X_test_scaled)
print("SMOTE 적용 전 분류 보고서:")
print(classification_report(y_test, predictions_before_smote))
print("SMOTE 적용 전 혼동 행렬:")
print(confusion_matrix(y_test, predictions_before_smote))

# SMOTE 적용
smote = SMOTE(sampling_strategy=0.5, k_neighbors=5, random_state=42)
X_sm, y_sm = smote.fit_resample(X_train_scaled, y_train)

# 로지스틱 회귀 모델 훈련 (SMOTE 적용 후)
model_after_smote = LogisticRegression(random_state=42, max_iter=1000)
model_after_smote.fit(X_sm, y_sm)

# 테스트 세트로 평가 (SMOTE 적용 후)
predictions_after_smote = model_after_smote.predict(X_test_scaled)
print("SMOTE 적용 후 분류 보고서:")
print(classification_report(y_test, predictions_after_smote))
print("SMOTE 적용 후 혼동 행렬:")
print(confusion_matrix(y_test, predictions_after_smote))

# 2차원 PCA 객체 생성
pca = PCA(n_components=2)

# 원본 훈련 데이터를 2차원으로 축소
X_train_pca = pca.fit_transform(X_train_scaled)

# SMOTE로 오버샘플링된 데이터를 2차원으로 축소
X_sm_pca = pca.transform(X_sm)

# 원본 데이터와 SMOTE로 오버샘플링된 데이터 시각화
fig, axes = plt.subplots(1, 2, figsize=(14, 7), sharex=True, sharey=True)

# 원본 데이터 시각화
axes[0].scatter(X_train_pca[y_train == 0, 0], X_train_pca[y_train == 0, 1], label=f'Class 0 (n={y_train.value_counts()[0]})')
axes[0].scatter(X_train_pca[y_train == 1, 0], X_train_pca[y_train == 1, 1], label=f'Class 1 (n={y_train.value_counts()[1]})')
axes[0].set_title('Original data')
axes[0].set_xlabel('PC1')
axes[0].set_ylabel('PC2')
axes[0].legend()

# SMOTE로 오버샘플링된 데이터 시각화
axes[1].scatter(X_sm_pca[y_sm == 0, 0], X_sm_pca[y_sm == 0, 1], label=f'Class 0 (n={pd.Series(y_sm).value_counts()[0]})')
axes[1].scatter(X_sm_pca[y_sm == 1, 0], X_sm_pca[y_sm == 1, 1], label=f'Class 1 (n={pd.Series(y_sm).value_counts()[1]})')
axes[1].set_title('SMOTE-based oversampled data')
axes[1].set_xlabel('PC1')
axes[1].set_ylabel('PC2')
axes[1].legend()

plt.tight_layout()
plt.show()
```
